### Method 1 - impute fixed value like 0 , 'Unknown ' or 'missing ' etc . we inpute unkown in gender columns.

## method 2 - impute Mean ,Median and mode

```python
# impute mean in amount _spent columns 
mean_amount_spent = df['Amount_spent'].mean()
df['Amount_spent'].fillna(mean_amount_spent, inplace=True)
df
mean_amount_spent
```

```python
# Impute median in age column
median_Age = df['Age'].median()
df['Age'].fillna(median_Age, inplace=True)
df
median_Age
```

```python
# impute mode in Employees_status columns
mode_emp = df['Employees_status'].mode().iloc[0]
df['Employees_status'].fillna(mode_emp, inplace=True)
mode_emp
```

# 4 Memory Management

#### when we on large datasets,there we get one big issue is a memory problem.We need too large resources for dealing with this . but there are same methodes in pands to dael with this here are same methods or strategies to deal with this problem with help of pands

```python
df_memory = df.copy()
```

```python
df_usage = df_memory.memory_usage(deep=True)
memory_usage_in_mbs = round(np.sum(memory_usage /1024 ** 2), 3)
print(f" Total memory taking df_memory dataframe is : {memory _usage_in_mbs:.2f} MB ")
```

## chnage oject to category datatypes

#### our data frame is samll in size . which is 1.15 MB now we will convert our object datatype to category

```python
# object datatype to category convert
df_memory[df_memory.select_dtypes(['object']).columns] = df_memory.select_dtype(['object']).apply(1ambda x: x.astype('category'))
df_memory
```

```python
# convert object to category
df_memory.info(memory_usage="deep")
```

# 5.3 Summarizing or grouping data


## groupby

### grouping by one columns : for eaxample ,if we want to find maximum values of age and amount _spent by gender then we can usen this

```python
df[['Age','Amount_spent']].groupby(df['Gender']).max()
```

### to find mean , count and max value of age and amount_spent by gender then we can use age() function with groupby()

```python
# Group by one columns
state_gender_res = df[['Age', 'Gender' , 'Amount_spent']].groupby(['Gender']).agg(['count','mean','max'])
state_gender_res
```

### grouping by multiple columns : to find total count,maximum and minimum values of Amount_spent by State_names, gender , and Payment _nethod then we can pass these columns names under groupby () function and add .agg() with count , mean, argument

```python
# Group by multiple columns
state_gender_res = df[['State_names','Gender','Payment_method','Amount_spent']].groupby([ 'State_names','Gender', 'Payment_method']).agg(['count', 'min', 'max'])
state_gender_res.head(12)
```
